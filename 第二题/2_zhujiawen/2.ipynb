## 图片分类
* 用卷积神经网络分类三种图片，可以使用深度学习库
* trainval为训练数据，文件名的首字母表示图片的真值类别
* test为测试数据
* 提交结果按图片名称的顺序将预测的类别标签存储为txt文件，示例见example.txt

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a_img = Image.open('trainval/A_1_000017.jpg')
a_img = np.array(a_img)

b_img = Image.open('trainval/B_2_000995.jpg')
b_img = np.array(b_img)

c_img = Image.open('trainval/C_1_002753.jpg')
c_img = np.array(c_img)

fig = plt.figure(figsize=(8, 8))
fig.add_subplot(1, 3, 1)
plt.imshow(a_img)
plt.xlabel('category: A')

fig.add_subplot(1, 3, 2)
plt.imshow(b_img)
plt.xlabel('category: B')

fig.add_subplot(1, 3, 3)
plt.imshow(c_img)
plt.xlabel('category: C')
```
